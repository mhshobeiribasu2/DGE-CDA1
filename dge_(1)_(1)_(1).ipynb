```python
import cv2
import json
import os

def resize_image(image_path, target_width=1280, target_height=720):
    # Read the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Failed to read {image_path}")
        return None

    # Resize the image
    resized_image = cv2.resize(image, (target_width, target_height))

    # Save the resized image
    cv2.imwrite(image_path, resized_image)
    return resized_image.shape[:2]  # Return new dimensions (height, width)

def update_annotations(annotations_path, new_dimensions):
    with open(annotations_path, 'r') as f:
        data = json.load(f)

    for annotation in data['annotations']:
        # Update the bounding box coordinates based on the new image dimensions
        bbox = annotation['bbox']
        x_scale = new_dimensions[1] / data['images'][0]['width']
        y_scale = new_dimensions[0] / data['images'][0]['height']
        annotation['bbox'] = [bbox[0] * x_scale, bbox[1] * y_scale, bbox[2] * x_scale, bbox[3] * y_scale]

    # Update image dimensions
    for image in data['images']:
        image['width'] = new_dimensions[1]
        image['height'] = new_dimensions[0]

    # Save the updated annotations
    with open(annotations_path, 'w') as f:
        json.dump(data, f, indent=4)


def process_dataset(images_dir, annotations_path, target_width=1280, target_height=720):
    with open(annotations_path, 'r') as f:
        data = json.load(f)

    for image_info in data['images']:
        image_path = os.path.join(images_dir, image_info['file_name'])
        height, width = image_info['height'], image_info['width']

        if width != target_width or height != target_height:
            print(f"Resizing {image_info['file_name']} from {width}x{height} to {target_width}x{target_height}")
            new_dimensions = resize_image(image_path, target_width, target_height)
            if new_dimensions:
                # Update image dimensions in the dataset
                image_info['width'], image_info['height'] = target_width, target_height

    # Update annotations
    update_annotations(annotations_path, (target_height, target_width))
    print("Dataset processing complete.")
images_dir = "DGE-CDA/givendata/imgs/"
annotations_path = "DGE-CDA/givendata/target.json"
process_dataset(images_dir, annotations_path, target_width=1280, target_height=720)
```
