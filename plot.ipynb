```python
import matplotlib.pyplot as plt

# Sample data
domain_gap_threshold = [0.005, 0.02, ]
total_energy_mBDD_DGE = [134380349, 130047955]
AP50_mBDD_DGE = [96,98]

# Plotting the data
fig, ax = plt.subplots()

ax.plot(domain_gap_threshold, total_energy_mBDD_DGE, 'go-', label='with DGE')

# Adding horizontal lines
ax.axhline(y=130356871, color='blue', linestyle='--', label='w/o DGE')

# Adding text annotations
for i, txt in enumerate(AP50_mBDD_DGE):
    ax.annotate(f'{txt}', (domain_gap_threshold[i], total_energy_mBDD_DGE[i]), textcoords="offset points", xytext=(0,10), ha='center', color='green')

# Setting labels
ax.set_xlabel('Domain gap threshold')
ax.set_ylabel('Total energy')

# Setting x-axis to log scale
ax.set_xscale('log')

# Adding the legend
ax.legend()

# Show plot
plt.show()
```
